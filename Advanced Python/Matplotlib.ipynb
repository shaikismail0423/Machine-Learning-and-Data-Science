## Pyplot
To start plotting graphs, we import the pyplot submodule from matplotlib. Pyplot is a collection of command style functions that make matplotlib work like MATLAB. 

Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

In order to plot a 2-D graph (say), we need to have a pair of coordinates: those of the x and y axes. 

If we declare our coordinates as arrays, the size of the x-coordinate array and y-coordinate array must be same. Matplotlib is convenient, in that in order to plot the coordinates with, say a scatter plot, 

we simply have to write pyplot.scatter() and we can have the points plotted on a scatter plot. Note that this will plot the graph but won’t display it;

we need to write pyplot.show() in order to display the graph

## Colors and Markers
You can change the colour of your plots by adding a ‘color’ parameter in the plot function definition (which can accept an arbitrary number of arguments, by the way).

Adding a ‘marker’ parameter will shape your plot points – say circular, triangular etc.

You need to assign the character ‘o’ for a circular shape and ‘^’ for triangular.

````python
# Evenly sampled time at 200ms intervals

t = np.arange(0., 5., 0.2)


# Red dashes, blue squares and green triangles
# We can plot multiple functions within one plot as well.

# First function: t vs. t^3
plt.plot(t, t**3, color='green', marker='^', label='t^3')

# Second function: t vs. t
plt.plot(t, t, color='blue', marker='o', label='t')

# Adding labels and legend
plt.xlabel('t')
plt.ylabel('y')
plt.title('Plot of t vs. t^3 and t')
plt.legend() #t  t^3

# Show the plot
plt.grid(True)
plt.show()
````

# Subplots
The subplot function allows one to plot multiple subplots within one plot. You need to specify the sizes which will be occupied by it in the bigger plot. For example, subplot(2, 3, 3) and subplot(233) both create an Axes at the top right corner of the current figure, occupying half of the figure height and a third of the figure width.

````python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1: (First subplot)
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# Creating the first subplot (2 rows, 1 column, and the 1st position)
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title('Plot 1')  # Adding a title to the first subplot

# Plot 2: (Second subplot)
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

# Creating the second subplot (2 rows, 1 column, and the 2nd position) layout
plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('Plot 2')  # Adding a title to the second subplot

plt.show()
````

# Different Variations for Graph Plotting
## Pie Charts
We next come to a fancier, more interesting way of representing data – pie graphs. Pie charts can be plotted very easily using the pie() function. We need to specify the sizes of each portion in the chart in a list.

Once you plot the pie chart, you might notice that the chart isn’t round – which can be remedied using the axis() function and passing it the parameter ‘equal’.

We’re done with our basic pie chart – how else can we customise it? We can pass it the ‘colors’ parameter and specify the colours of each section, and also give each section labels using the ‘labels’ parameter.

How do we display information on the chart itself? We can display the percentage of each section of the pie by giving the parameter ‘autopct’. The autopct parameter needs a function (in the form of a string) that demonstrates how to display the percentage on the chart – for instance, ‘%.2f%%’ will display the percentages with two digits to the right of decimal. And it is not necessary that just the percentages can be displayed on the pie chart; we can pass any function we wish into the autopct parameter and it’ll decide the values on its own.

You must have seen in several places that sometimes, for more interesting insight, a certain part of the pie chart is usually shown to be slightly out of the pie – a way of highlighting it. We can do that by passing the ‘explode’ parameter and giving it the values by which each section should be pushed out of the pie chart. A value of 1 indicated that the section is to be pushed entirely out of the pie.

There are many, many other parameters we can pass to the pie() function to make our graph even better and informative. For instance, the counterclock parameter can be used to specify fractions direction, clockwise or counter-clockwise. The wedgeprops parameter takes in a dictionary of arguments passed to the wedge objects making the pie. For example, you can pass in wedgeprops = {'linewidth': 3} to set the width of the wedge border lines equal to 3.

````python
import matplotlib.pyplot as plt

labels = ["A","B","C","D"]
sizes = [3, 4, 6, 2]
colors = ["blue","red","pink","purple"]
plt.title("Split amound classes")
plt.pie(sizes, colors = colors,labels= labels, autopct="%0.2f%%",
        counterclock=False,
       startangle= 90) 
#"% -> it indicates some special charcer
plt.axis("equal")
plt.show()
````

````python
import matplotlib.pyplot as plt

# Data for the pie chart
pieLabels = ['Asia', 'Africa', 'Europe', 'North America', 'South America']
populationShare = [59.69, 16, 9.94, 7.79, 5.68]

# Create a figure and axes object
figureObject, axesObject = plt.subplots()

# Plot the pie chart
# 'labels' - the labels for each section of the pie chart
# 'autopct' - format for displaying the percentage values in each section
# 'startangle' - angle at which the first wedge starts (90 degrees for a vertical start)
# 'wedgeprops' - properties of the wedges, like linewidth
axesObject.pie(populationShare, labels=pieLabels, autopct='%.1f%%',
               startangle=90, wedgeprops={'linewidth': 5})

# Set the aspect ratio of the pie chart to be equal so that it appears as a circle
axesObject.axis('equal')

# Show the pie chart
plt.show()
````

````python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  
# only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
company = ['HP','Dell','Lenovo','Asus','Apple','Acer']
sold = [20000,43000,15000,17000,22000,13000]

company = np.array(company)
sold = np.array(sold)

plt.pie(sold, labels=company, autopct='%1.1f%%', counterclock=False)
plt.axis('equal')
plt.show()

# Calculate the sum of 'sold' along the specified axis (axis=0, i.e., column-wise sum)
# keepdims=True ensures the result has the same dimensions as the original array
sold_sum = sold.sum(axis=0, keepdims=True)
print(sold_sum)

# Perform element-wise division of 'sold' by 'sold_sum' and multiply by 100 to get percentages
x = np.true_divide(sold, sold_sum) * 100

for i in range(len(company)):
    print(company[i],format(x[i],'.1f'))
````

# Bubble Plots
Bubble plots are an improved version of the scatter plot. In a scatter plot, there are two dimensions x, and y. In a bubble plot, there are three dimensions x, y, and z. Where the third dimension z denotes weight. That way, bubble plots give more information visually than a two dimensional scatter plot.

````python
import numpy as np
import matplotlib.pyplot as plt

N = 50
x = np.random.rand(N)    # Generate N random values between 0 and 1 for x
y = np.random.rand(N)    # Generate N random values between 0 and 1 for y
colors = np.random.rand(N)    # Generate N random values between 0 and 1 for colors

# Generate random values for the area of points (s)
area = np.pi * (15 * np.random.rand(N))**2

# Create the scatter plot
# 'x' and 'y' - arrays of x and y coordinates of data points
# 's' - size of points, can be a single number or an array to specify size for each point
# 'c' - color of the points, can be a single color or an array to specify color for each point
# 'alpha' - transparency value of colors, ranging from 0 (transparent) to 1 (opaque)

plt.scatter(x, y, s=area, c=colors, alpha=0.5)

# Show the plot
plt.show()
````

If bins is an integer, it defines the number of equal-width bins in the range.

If bins is a sequence, it defines the bin edges, including the left edge of the first bin and the right edge of the last bin; in this case, bins may be unequally spaced. All but the last (righthand-most) bin is half-open. In other words, if bins is:

[1, 2, 3, 4]

then the first bin is [1, 2) (including 1, but excluding 2) and the second [2, 3). The last bin, however, is [3, 4], which includes 4.

If bins is a string, it is one of the binning strategies supported by numpy.histogram_bin_edges: 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', or 'sqrt'.

````python
import csv
import matplotlib.pyplot as plt
import collections

with open('amazon_jobs_dataset.csv', encoding ='UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)

    dct = {}
    for row in file_data:
        date = row['Posting_date'].split(",")
        key = date[1]

        if key in dct:
            dct[key] += 1
        else:
            dct[key] = 1
            
    ord_list_of_tuples = sorted(dct.items())  
    ord_dct = collections.OrderedDict(ord_list_of_tuples)
    plt.plot(list(ord_dct.keys()),list(ord_dct.values()))
    plt.xlabel("Year")
    plt.ylabel("Jobs")
    plt.title('Year vs Job_Openings')
    plt.show()
    
    for i in ord_dct.keys():
        print(i,end=" ")
        print(ord_dct[i])
````

Deciding which graph type to use depends on the nature of your data and the insights you want to convey. Here's a brief overview of each graph type and when to use them:

1. Line:
   - Use when you want to show trends over time or sequential data points.
   - Suitable for continuous data, where the order matters.
   - Can be used to visualize time series data, stock prices, temperature trends, etc.

2. Scatter:
   - Use when you want to visualize the relationship between two continuous variables.
   - Each data point represents an individual observation.
   - Suitable for detecting patterns, correlations, or outliers in the data.

3. Bar:
   - Use when you want to compare different categories or discrete groups.
   - Suitable for displaying data with distinct categories or nominal data.
   - Useful for comparing quantities, frequencies, or counts across categories.

4. Pie:
   - Use when you want to show the proportion or percentage distribution of parts to a whole.
   - Suitable for representing parts of a whole, such as market shares, survey responses, or demographic percentages.
   - Avoid using pie charts for more than a few categories, as it can become hard to interpret.

5. Bubble:
   - Use when you want to visualize three dimensions of data on a two-dimensional plot.
   - Similar to scatter plots, but you can add a third dimension using the size of the data points (bubbles).
   - Suitable for showing relationships among three variables simultaneously.

6. Histogram:
   - Use when you want to display the distribution of continuous or discrete data.
   - Suitable for visualizing the frequency or count of data in different bins or intervals.
   - Helps identify the central tendency, spread, and skewness of the data.

Keep in mind that the choice of graph type should align with the objectives of your analysis and the characteristics of the data you are working with. Experiment with different graph types to see which one best represents your data and effectively communicates your insights.
