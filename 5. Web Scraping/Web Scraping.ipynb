```python
import requests
from bs4 import BeautifulSoup

print("Started")

base_url = "https://books.toscrape.com/catalogue/"
books_urls = []

page = "page-1.html"
current_page = base_url + page

while True:
    response = requests.get(current_page)
    
    if response.status_code != 200:
        break
    
    data = BeautifulSoup(response.text, "html.parser")
    li = data.find_all(class_="product_pod")
    
    for book in li:
        books_urls.append("https://books.toscrape.com/" + book.h3.a['href'])
    
    next_ele = data.find(class_="next")
    
    if next_ele is None:
        break
    else:
        page = next_ele.a['href']
        current_page = base_url + page

print("Completed")
```
