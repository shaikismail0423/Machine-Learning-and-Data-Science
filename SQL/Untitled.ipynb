MySQL is a popular relational database management system that uses SQL (Structured Query Language) to interact with databases. Below are some commonly used MySQL commands:

1. **Connect to MySQL server:**
   ```
   mysql -u username -p
   ```
   Replace `username` with your MySQL username. The `-p` flag will prompt you to enter your password.

2. **Show available databases:**
   ```
   SHOW DATABASES;
   ```

3. **Create a new database:**
   ```
   CREATE DATABASE database_name;
   ```
   Replace `database_name` with the name you want to give to the new database.

4. **Switch to a specific database:**
   ```
   USE database_name;
   ```
   Replace `database_name` with the name of the database you want to use.

5. **Show tables in the current database:**
   ```
   SHOW TABLES;
   ```

6. **Create a new table:**
   ```
   CREATE TABLE table_name (
       column1 datatype constraint,
       column2 datatype constraint,
       ...
   );
   ```
   Replace `table_name`, `column1`, `column2`, etc. with the desired table and column names, and `datatype` with the appropriate data type for each column. You can also add constraints such as `PRIMARY KEY`, `FOREIGN KEY`, `NOT NULL`, etc.

7. **Insert data into a table:**
   ```
   INSERT INTO table_name (column1, column2, ...) VALUES (value1, value2, ...);
   ```
   Replace `table_name`, `column1`, `column2`, etc. with the appropriate table and column names, and `value1`, `value2`, etc. with the values you want to insert.

8. **Select data from a table:**
   ```
   SELECT column1, column2, ... FROM table_name WHERE condition;
   ```
   Replace `column1`, `column2`, etc. with the columns you want to retrieve, `table_name` with the table you want to query, and `condition` with any filtering criteria you want to apply.

9. **Update data in a table:**
   ```
   UPDATE table_name SET column1 = new_value1, column2 = new_value2 WHERE condition;
   ```
   Replace `table_name`, `column1`, `column2`, etc. with the appropriate table and column names, `new_value1`, `new_value2`, etc. with the new values you want to set, and `condition` with the filtering criteria to determine which rows to update.

10. **Delete data from a table:**
    ```
    DELETE FROM table_name WHERE condition;
    ```
    Replace `table_name` with the table you want to delete data from, and `condition` with the filtering criteria to determine which rows to delete.

11. **Drop a table:**
    ```
    DROP TABLE table_name;
    ```
    Replace `table_name` with the name of the table you want to drop (permanently delete).

These are some of the fundamental MySQL commands that will help you interact with databases effectively. Remember to be cautious while executing commands as they can have a significant impact on the data stored in your database. Always take backups and double-check your queries before running them, especially in a production environment.
